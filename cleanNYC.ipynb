```python
def swap_columns(data, col1, col2):
    col_list = list(data.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    data = data[col_list]
    return data
```

```python
data[['day','M','D','time','Timezone','Y']] = data['UTC Time'].str.split(' ',expand = True)

d = dict(zip(pd.date_range('2000-01-01', freq='M', periods=12).strftime('%b'), ["%02d" % i for i in range(1,13)]))
data.M = data.M.map(d)
data.drop(columns=['UTC Time', 'Timezone'],inplace = True)
data = swap_columns(data, 'Y', 'time')
data = swap_columns(data, 'D', 'M')
data.head()
```

```python
print("longitude min: ", min(data.iloc[:,1]))
print("longitude max: ", max(data.iloc[:,1]))
print("latitude min: ", min(data.iloc[:,0]))
print("latitude max: ", max(data.iloc[:,0]))
```

```python
data['Timestamp']=data[['Y','M','D']].astype(str).agg('-'.join, axis=1)
data['Timestamp']=pd.to_datetime(data['Timestamp'] + ' ' + data['time'], utc = True)

data.head()
```

```python
data['ts'] = data['Timestamp'].apply(lambda x : x.astype(int)/(10**9)).map('{:.0f}'.format)
data.drop(columns=['day','D','M','Y','time','Timestamp'],inplace = True)
data.dtypes
```
